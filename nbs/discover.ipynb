# Discover

Create some graphs to discover the data.

```python
from datetime import datetime

import numpy as np
import pandas as pd

from pyzonnestand import observed_sun_position

# every hour for the last year
dt = np.arange(
    np.datetime64("2019-01-01"), np.datetime64("2020-01-01"), np.timedelta64(1, "h")
)

python_dt = dt.astype(datetime)
```

```python
x = observed_sun_position(
    dt=python_dt,
    latitude=52.0,
    longitude=5.0,
    elevation=0.0,
)
```

```python
df = pd.DataFrame(x).T
df.columns = ["azimuth", "zenith"]
df["elevation"] = 90 - df["zenith"]
df.index = dt

df.head()
```

```python
df.loc["2019-01", "elevation"].plot(figsize=(20, 4))
```
